#### Q5. Perform parameter selection for Lasso and Ridge over validation sample
try exponential range -exp(10)...exp(20) with 100 values in it

Visualize the validation R2 vs alpha

For the optimal alpha evaluate performance of the models over the test set

```python
def Regularization_fit_alpha(model,X_train,y_train,X_valid,y_valid,alphas,Graph=False, logl=False):

    R_2_OS=[] 
    if model==1:
        RM = lambda a: linear_model.Ridge(fit_intercept=True, alpha=a)
        model_label='Ridge'
    else:
        RM = lambda a: linear_model.Lasso(fit_intercept=True, alpha=a)
        model_label='Lasso'
    
    best_R2 = -1
    best_alpha = alphas[0]
    best_coefs = []
    
    for a in alphas: 
        lm = RM(a) 
        lm.fit(X_train,y_train)  
        y_predict=lm.predict(X_valid) 
        R_2_OS_=r2_score(y_valid,y_predict)
        R_2_OS.append(R_2_OS_)
        if R_2_OS_ > best_R2: 
            best_R2 = R_2_OS_
            best_alpha = a
            best_coefs = lm.coef_
    
    if Graph==True: 
        plt.figure(figsize = (15,10))
        plt.title('OS-R-squared for different Alpha')
        if logl: 
            plt.xlabel('ln(Alpha)')
            l=np.log(alphas)
            bl=np.log(best_alpha)
        else:
            plt.xlabel('Alpha')
            l=alphas
            bl=best_alpha
        plt.ylim((-0.1,1))
        plt.plot(l,R_2_OS,'b',label=model_label)
        plt.legend(loc='upper right')
        plt.ylabel('R-squared')
        plt.axvline(bl,color='r',linestyle='--')

        plt.show()
    
    return [best_alpha,best_R2,best_coefs]
```

```python
alphas = np.exp(np.linspace(-10.0,20.0,100)) 
for x in range(1,3):
  if x ==1:
    [alpha_optimalR, best_R2R, best_coefsR]=Regularization_fit_alpha(x,al,REsample1_311['sale_price'],ag,REsample2_311['sale_price'],alphas,Graph=True,logl=True)
    print('Optimal alpha for Ridge={0}, best OS R_2={1}'.format(alpha_optimalR,best_R2R))
  else:
    [alpha_optimalL, best_R2L, best_coefsL]=Regularization_fit_alpha(x,al,REsample1_311['sale_price'],ag,REsample2_311['sale_price'],alphas,Graph=True,logl=True)
    print('Optimal alpha for Lasso={0}, best OS R_2={1}'.format(alpha_optimalL,best_R2L))
```

```python
Lasso=linear_model.Lasso(fit_intercept=True, alpha=alpha_optimalL)
Lasso.fit(al,REsample1_311['sale_price'])

print("The IS R-squared of Lasso is: {0}".format(r2_score(REsample1_311['sale_price'],Lasso.predict(al))))
print("The OS R-squared of Lasso is: {0}".format(r2_score(REsample3_311['sale_price'],Lasso.predict(ab))))
```

```python
Ridge=linear_model.Ridge(fit_intercept=True, alpha=alpha_optimalR)
Ridge.fit(al,REsample1_311['sale_price'])

print("R-squared of Ridge for Training is: {}".format(r2_score(REsample1_311['sale_price'],Ridge.predict(al))))
print("R-squared of Ridge for Testing is: {}".format(r2_score(REsample3_311['sale_price'],Ridge.predict(ab))))
```
